```python
import pandas as pd

class IrisDataIterator:
    def __init__(self, filename):
        self.data = pd.read_csv(filename)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            print("Error: Los datos se han agotado.")
            raise StopIteration
        else:
            row = self.data.iloc[self.index]
            self.index += 2
            return row[0], row[1]


iris_iterator = IrisDataIterator("iris.csv")
for _ in range(1000000):
    try:
        data = next(iris_iterator)
        print("Dato 1:", data[0])
        print("Dato 2:", data[1])
    except StopIteration as e:
        print(e)
        break
```
